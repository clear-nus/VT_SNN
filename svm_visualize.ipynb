```python
import pickle
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import IFrame, display
```

```python
lengths1, cv_std1, cv_mean1, test_accs1, C_params1 = pickle.load(open('svm_results_1.pkl', 'rb'))
lengths2, cv_std2, cv_mean2, test_accs2, C_params2 = pickle.load(open('svm_results_2.pkl', 'rb'))
lengths3, cv_std3, cv_mean3, test_accs3, C_params3 = pickle.load(open('svm_results_3.pkl', 'rb'))
lengths4, cv_std4, cv_mean4, test_accs4, C_params4 = pickle.load(open('svm_results_4.pkl', 'rb'))
lengths5, cv_std5, cv_mean5, test_accs5, C_params5 = pickle.load(open('svm_results_5.pkl', 'rb'))
```

```python
lengths = lengths1 + lengths2 + lengths3 + lengths4 + lengths5
cv_std = cv_std1 + cv_std2 + cv_std3 + cv_std4 + cv_std5
cv_mean = cv_mean1 + cv_mean2 + cv_mean3 + cv_mean4 + cv_mean5
test_accs = test_accs1 + test_accs2  + test_accs3 + test_accs4 + test_accs5
C_params = C_params + C_params2 + C_params3 + C_params4 + C_params5
```

```python
df = pd.DataFrame(columns=['timestamp', 'cv_std', 'cv_mean', 'test_accs', 'C_params'])
df.timestamp = np.array(lengths)/50
df.cv_std = cv_std
df.cv_mean = cv_mean
df.test_accs = test_accs
df.C_params = C_params
```

```python
filepath = "object_det_over_time.pdf"
IFrame(filepath, width=700, height=400)
```

```python
fig, ax = plt.subplots(2, figsize=(10,6))
sns.lineplot(ax=ax[0], x="timestamp", y="test_accs", data=df)
sns.lineplot(ax=ax[1], x="timestamp", y="C_params", data=df)
plt.show()
```
