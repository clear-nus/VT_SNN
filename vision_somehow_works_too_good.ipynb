```python
from datetime import datetime
import matplotlib.pyplot as plt
from torch.utils.data import Dataset, DataLoader
import zipfile
import torch
from torch import nn
import numpy as np
import copy
from pathlib import Path

ref_name = 'models_and_stats2/mlp_lstm_vision_0'
device = torch.device('cuda:2')
num_epochs = 2000
```

```python
# Dataset definition
class ViTacDataset(Dataset):
    def __init__(self, datasetPath, sampleFile):
        self.path = datasetPath 
        self.samples = np.loadtxt(sampleFile).astype('int')
        self.vis = torch.load(Path(self.path) / "ds_vis.pt")

    def __getitem__(self, index):
        inputIndex  = self.samples[index, 0]
        classLabel  = self.samples[index, 1]
        return self.vis[inputIndex], classLabel

    def __len__(self):
        return self.samples.shape[0]
```

```python
# Dataset and dataLoader instances.
split_list = ['80_20_1','80_20_2','80_20_3','80_20_4','80_20_5']
```

```python
data_dir = '/home/tasbolat/some_python_examples/data_VT_SNN/'
pooled_viz_dir = data_dir + "pooled_vis/"


sample_file = 1
trainingSet = ViTacDataset(datasetPath = pooled_viz_dir, sampleFile = data_dir + "/train_" + split_list[sample_file-1] + ".txt")
trainLoader = DataLoader(dataset=trainingSet, batch_size=8, shuffle=False, num_workers=8)
 
testingSet = ViTacDataset(datasetPath = pooled_viz_dir, sampleFile  = data_dir + "/test_" + split_list[sample_file-1] + ".txt")
testLoader = DataLoader(dataset=testingSet, batch_size=8, shuffle=False, num_workers=8)
```

```python
class MultiMLP_LSTM(nn.Module):

    def __init__(self):
        super(MultiMLP_LSTM, self).__init__()
        self.seq_length = 325
        self.hidden_dim = 30
        self.batch_size = 8
        self.num_layers = 1

        # Define the LSTM layer
        self.lstm = nn.GRU(self.seq_length, self.hidden_dim, self.num_layers)

        # Define the output layer
        self.fc = nn.Linear(self.hidden_dim, 20)
        
        #self.pool_vis = nn.AvgPool3d((1,5,5), padding=0, stride=(1,7,7))
        self.fc_vis = nn.Linear(63*50*2, 10)


    def forward(self, in_vis):

        in_vis = in_vis.reshape([in_vis.shape[0], 325, 50*63*2])
        #print('in vis:', in_vis.shape)
        lstm_vis = self.fc_vis(in_vis).permute(0,2,1)
        #print('lstm in:', lstm_vis.shape)
        #print(lstm_vis.shape, lstm_tac.shape)
        #lstm_in = torch.cat([lstm_tac, lstm_vis], dim=2).permute(0,2,1)
#         print(lstm_in.shape)
                
        lstm_out, self.hidden = self.lstm(lstm_vis)
        
        # Only take the output from the final timetep
        # Can pass on the entirety of lstm_out to the next layer if it is a seq2seq prediction
        y_pred = self.fc(lstm_out[:, -1, :])
        
        return y_pred
```

```python
train_accs = []
test_accs = []
train_loss = []
test_loss = []
```

```python
net = MultiMLP_LSTM().to(device)
# Create snn loss instance.
criterion = nn.CrossEntropyLoss()
# Define optimizer module.
optimizer = torch.optim.RMSprop(net.parameters(), lr = 0.0001)
```

```python
for epoch in range(1, num_epochs+1):
    # Training loop.
    net.train()
    correct = 0
    batch_loss = 0
    train_acc = 0
    for i, (in_vis, label) in enumerate(trainLoader, 0):

        in_vis = in_vis.to(device)
        #in_vis = in_vis.squeeze().permute(0,1,4,2,3)
        #print(in_vis.shape)
        label = label.to(device)
        # Forward pass of the network.
        out_tact = net.forward(in_vis)
        # Calculate loss.
        loss = criterion(out_tact, label)
        #print(loss)

        batch_loss += loss.cpu().data.item()
        # Reset gradients to zero.
        optimizer.zero_grad()
        # Backward pass of the network.
        loss.backward()
        # Update weights.
        optimizer.step()

        _, predicted = torch.max(out_tact.data, 1)
        correct += (predicted == label).sum().item()

    # Reset training stats.
    train_acc = correct/len(trainLoader.dataset)
    train_loss.append(batch_loss/len(testLoader.dataset))
    train_accs.append(train_acc)
    #print(train_acc, batch_loss)

    # testing
    net.eval()
    correct = 0
    batch_loss = 0
    test_acc = 0
    with torch.no_grad():
        for i, (in_vis, label) in enumerate(testLoader, 0):
            in_vis = in_vis.to(device)
            #in_vis = in_vis.squeeze().permute(0,1,4,2,3)

            # Forward pass of the network.
            out_tact = net.forward(in_vis)
            label = label.to(device)
            _, predicted = torch.max(out_tact.data, 1)
            correct += (predicted == label).sum().item()
            # Calculate loss.
            loss = criterion(out_tact, label)
            batch_loss += loss.cpu().data.item()

    test_loss.append(batch_loss/len(testLoader.dataset))
    test_acc = correct/len(testLoader.dataset)
    test_accs.append(test_acc)
#     if epoch % 50 ==0:
#         print('------------------------')
#         print('saving model')
#         torch.save(net.state_dict(), ref_name + '_' + str(epoch) + '_' + str(k) + ".pt")
#         print('Train:', train_acc, 'Test:', test_acc)
#         print('------------------------')
    if epoch%1 == 0:
        print(epoch, 'Train:', train_acc, 'Test:', test_acc)
```

```python
fig, ax = plt.subplots(2, figsize=(15,7))
ax[0].plot(train_loss)
ax[0].plot(test_loss)
ax[0].legend(['Train', 'Test'])
ax[0].set_xlabel('Epochs')
ax[0].set_ylabel('Loss')

ax[1].plot(train_accs)
ax[1].plot(test_accs)
ax[1].legend(['Train', 'Test'])
ax[1].set_xlabel('Epochs')
ax[1].set_ylabel('Accuracy')

plt.show()
```
