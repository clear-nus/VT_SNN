```python
# IMPORTANT!

# apply only if needed to remove "outlier", brings down number of samples to 15 per each class
# we identify "outliers manually" - by spiketrain distance
remove_outlier = True
```

```python
import numpy as np
from sklearn.model_selection import StratifiedKFold
np.random.seed(0)
```

```python
remove_list = np.loadtxt('blacklisted.txt').astype(int)
```

```python
# batch2
list_of_objects2 = [
    '107-a_pepsi_bottle',
    '107-b_pepsi_bottle',
    '107-c_pepsi_bottle',
    '107-d_pepsi_bottle',
    '107-e_pepsi_bottle',
    '108-a_tuna_fish_can',
    '108-b_tuna_fish_can',
    '108-c_tuna_fish_can',
    '108-d_tuna_fish_can',
    '108-e_tuna_fish_can',
    '109-a_soymilk',
    '109-b_soymilk',
    '109-c_soymilk',
    '109-d_soymilk',
    '109-e_soymilk',
    '110-a_coffee_can',
    '110-b_coffee_can',
    '110-c_coffee_can',
    '110-d_coffee_can',
    '110-e_coffee_can'
]
```

```python
test_labels = {'half_split':[], 
              '80_20_1':[],
              '80_20_2':[],
              '80_20_3':[],
              '80_20_4':[],
              '80_20_5':[]}
train_labels = {'half_split':[], 
              '80_20_1':[],
              '80_20_2':[],
              '80_20_3':[],
              '80_20_4':[],
              '80_20_5':[]}
```

```python
labels = []
current_label = -1
overall_count = -1
for obj in list_of_objects2:
    current_label += 1
    for i in range(0, 20):
        overall_count+=1
        if remove_outlier:
            if overall_count in remove_list[:,0]:
                continue
        labels.append([overall_count, current_label])
labels = np.array(labels)
```

```python
skf = StratifiedKFold(n_splits=5, random_state=100, shuffle=True)
train_indices = []
test_indices = []
for train_index, test_index in  skf.split(np.zeros(len(labels)), labels[:,1]):
    train_indices.append(train_index)
    test_indices.append(test_index)
    print("TRAIN:", train_index, "TEST:", test_index)
```

```python
splits = ['80_20_1','80_20_2','80_20_3','80_20_4', '80_20_5']
```

```python
count = 0
for split in splits:
    np.savetxt('splits/' + 'train_' + split + '.txt', np.array(train_indices[count], dtype=int), fmt='%d', delimiter='\t')
    np.savetxt('splits/' + 'test_' + split + '.txt', np.array(test_indices[count], dtype=int), fmt='%d', delimiter='\t')
```
