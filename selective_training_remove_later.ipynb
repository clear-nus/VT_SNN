```python
import numpy as np
```

```python
# batch2
list_of_objects2 = [
    "107-a_pepsi_bottle",
    "107-b_pepsi_bottle",
    "107-c_pepsi_bottle",
    "107-d_pepsi_bottle",
    "107-e_pepsi_bottle",
    "108-a_tuna_fish_can",
    "108-b_tuna_fish_can",
    "108-c_tuna_fish_can",
    "108-d_tuna_fish_can",
    "108-e_tuna_fish_can",
    "109-a_soymilk",
    "109-b_soymilk",
    "109-c_soymilk",
    "109-d_soymilk",
    "109-e_soymilk",
    "110-a_coffee_can",
    "110-b_coffee_can",
    "110-c_coffee_can",
    "110-d_coffee_can",
    "110-e_coffee_can",
]
```

```python
# create labels
labels = []
current_label = -1
overall_count = 14

e = 1

for obj in list_of_objects2:
    current_label += 1
    
    for i in range(15, 15+20, 1):
        overall_count+=1
        labels.append([overall_count, current_label])
    overall_count += 20
    e+=1
labels = np.array(labels)
```

```python
labels
```

```python
from sklearn.model_selection import StratifiedKFold
```

```python
# stratified k fold
skf = StratifiedKFold(n_splits=5, random_state=100, shuffle=True)
train_indices = []
test_indices = []
```

```python
for train_index, test_index in  skf.split(np.zeros(len(labels)), labels[:,1]):
    train_indices.append(train_index)
    test_indices.append(test_index)
    print("TRAIN:", train_index, "TEST:", test_index)
```

```python
save_dir = '../data_VT_SNN_2/'
```

```python
# write to the file
splits = ['80_20_6','80_20_7','80_20_8','80_20_9', '80_20_10']
count = 0
for split in splits:
    tr = labels[train_indices[count], :]
    np.random.shuffle(tr)
    np.savetxt(save_dir + 'train_' + split + '.txt', np.array(tr, dtype=int), fmt='%d', delimiter='\t')
    np.savetxt(save_dir + 'test_' + split + '.txt', np.array(labels[test_indices[count], :], dtype=int), fmt='%d', delimiter='\t')
    count += 1
```
