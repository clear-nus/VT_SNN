```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pickle
import os.path
```

# Tactile

```python
models = ['tact', 'vis']
lrs = [0.1, 0.5, 0.01, 0.05, 0.001, 0.005, 0.0001, 0.0005]
batch_sizes=[1,8,64,128,256]
```

```python
all_info = []
for k in range(2):
    for lr in lrs:
        for batch_size in batch_sizes:
            fname = 'stats/' + models[k] +str(batch_size) +'_'+ str(lr) +'.pk'
            if os.path.isfile(fname):
                
                [tr_losses, te_losses, tr_accs, te_accs] = pickle.load(open(fname, 'rb'))
                vals = [models[0], batch_size, lr,
                          np.max(tr_accs), tr_accs[-1], tr_losses[-1], np.min(tr_losses),
                          np.max(te_accs), te_accs[-1], te_losses[-1], np.min(te_losses)]
                all_info.append(vals)
```

```python
column_names = ['Model', 'batch_size', 'lr', 
               'max_train_acc', 'train_acc', 'train_loss', 'min_train_loss',
              'max_test_acc', 'test_acc', 'test_loss', 'min_test_loss']
df = pd.DataFrame(data=np.array(all_info), columns=column_names)
```
