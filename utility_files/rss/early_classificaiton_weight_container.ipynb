```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib.legend import Legend
import matplotlib.pyplot as plt
```

```python
sns.set_context('paper')
sns.set_style('whitegrid')
sns.set(rc={"xtick.bottom" : True, 'font.sans-serif': 'Liberation Sans'}, context='paper', style='whitegrid', palette='Set1')
```

```python
df = pd.read_csv('early_class_obj_class.csv')
```

```python
df.head()
```

```python
df_spike_count = df[df['Loss function'] == 'Spike Count'].drop(columns=['Loss function'], axis=1)
df_weighted_spike_count = df[df['Loss function'] == 'Weighted Spike Count'].drop(columns=['Loss function'], axis=1)
```

```python
fig, ax = plt.subplots(figsize=(6,4))
sns.lineplot(ax = ax, x='t', y='Accuracy', hue='Modality', data = df_spike_count, ci='sd', linewidth=2, hue_order=['Tactile','Vision', 'Combined'], alpha=0.3, markers='--')
ax.lines[0].set_linestyle("--")
ax.lines[1].set_linestyle("--")
ax.lines[2].set_linestyle("--")

sns.lineplot(ax = ax, x='t', y='Accuracy', hue='Modality', data = df_weighted_spike_count, ci='sd', linewidth=2, hue_order=['Tactile','Vision', 'Combined'])

handles, labels = ax.get_legend_handles_labels()
ax.legend_.remove()

# leg = Legend(ax, handles=handles[1:4], labels=labels[1:4], loc='lower right', frameon=True, prop={'size': 18})
# ax.add_artist(leg)

leg = Legend(ax, handles=handles[1:4], labels=labels[1:4], loc='upper left', frameon=True, prop={'size': 14})
ax.add_artist(leg)
# leg = Legend(ax, handles=handles[5:], labels=labels[5:],loc='lower right', frameon=True, prop={'size': 14})
# ax.add_artist(leg)


ax.set_xticks(np.arange(0,6.5,0.25), minor=True)
ax.tick_params(axis='x', which='minor')
ax.set_xlim([0,6.5])
ax.set_ylim([0,1])

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('Time (s)', fontsize=18)
plt.ylabel('Test Accuracy', fontsize=18)
plt.tight_layout(True)
plt.savefig('early_class_obj_Lw2.svg', dpi=300)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(6,4))
ax = sns.lineplot(x='t', y='Accuracy', hue='Modality', data = df_spike_count, ci='sd', linewidth=2, hue_order=['Tactile','Vision', 'Combined'])
handles, labels = ax.get_legend_handles_labels()
ax.legend_.remove()

leg = Legend(ax, handles=handles[1:4], labels=labels[1:4], loc='upper left', frameon=True, prop={'size': 18})
ax.add_artist(leg)

ax.set_xticks(np.arange(0,6.5,0.25), minor=True)
ax.tick_params(axis='x', which='minor')
ax.set_xlim([0,6.5])
ax.set_ylim([0,1])

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('Time (s)', fontsize=18)
plt.ylabel('Test Accuracy', fontsize=18)
plt.tight_layout(True)
plt.savefig('early_class_obj_L.svg', dpi=300)
plt.show()
```
