```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

```python
x = np.arange(0, 2*np.pi, 0.01)

start_function = np.sin(x)
target_function = np.cos(x)

N = 100
diff = (target_function - start_function)/N
interpolated_function = start_function + diff*30
```

```python
fig, ax = plt.subplots()

line, = ax.plot(x, start_function, 'b')
line, = ax.plot(x, target_function, 'r')
line, = ax.plot(x, interpolated_function, 'g')
#plt.show()
```

```python
N = 10 # number of points
```

```python
def init():  # only required for blitting to give a clean slate.
    line.set_ydata(start_function)
    #line2.set_ydata(target_function, 'r')

    return line


def animate(i):
    #line.set_ydata(np.sin(x + i / 100))  # update the data.
    i = i%100
    interpolated_function = start_function + diff*i
    line.set_ydata(interpolated_function)
    return line,


ani = animation.FuncAnimation(
    fig, animate, init_func=init, interval=2, blit=True, save_count=50)
```
