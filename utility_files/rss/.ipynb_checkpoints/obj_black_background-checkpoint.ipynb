```python
#pepsi, tuna, soy, coffee
list_of_items = {
    'pepsi':[0, 20, 40, 60, 80],
    'tuna':[100, 120, 140, 160, 180],
    'soy':[200, 220, 240, 260, 280],
    'coffee':[300, 320, 340, 360, 380],
}
mappings = {'a':0, 'b':1, 'c':2, 'd':3, 'e':4}


iter_num = list_of_items[args.obj][mappings[args.weight_class]]

# # please indicate data dir
# data_dir = '/home/tasbolat/some_python_examples/data_VT_SNN_new/'
train_dataset = torch.load(Path(args.data_dir) / "ds_vis.pt")
```

```python
def get_data(num):
    # this function outputs input tactile, vision and downsampled vision data
    vis = torch.FloatTensor(np.load(Path(args.data_dir) / f'{num:d}_vis.npy'))
    tac = torch.FloatTensor(np.load(Path(args.data_dir) / f'{num:d}_tact.npy'))
    return tac, vis, train_dataset[num]

### Generate for each object
tac, vis, ds_vis = get_data(iter_num)
#ttac = tac.reshape([-1,325])
```

```python
ttac=tac[:39, :, :] 
```

```python
ttac.shape
```

```python
def get_plottable_st(spike_input):
    spike_trains = []
    bbb = []
    for n in range(spike_input.shape[0]):
        bbb.append([])
    for t in range(325):
        for n in range(spike_input.shape[0]):
            if spike_input[n,t] != 0:
                bbb[n] += [t*0.02]
        spike_trains.append(copy.deepcopy(bbb))
    return spike_trains
```

```python
def get_plottable_st_neg(spike_input):
    spike_trains = []
    bbb = []
    for n in range(spike_input.shape[0]*2):
        bbb.append([])
    for t in range(325):
        for n in range(spike_input.shape[0]):
            if spike_input[n,t] != 0:
                bbb[n+spike_input.shape[0]] += [t*0.02]
        spike_trains.append(copy.deepcopy(bbb))
    return spike_trains
```

```python
spike_trains_pos = get_plottable_st(ttac[:, 0, :])
spike_trains_neg = get_plottable_st_neg(ttac[:, 1, :])
```

```python
fig = plt.figure(figsize=(10,5))
plt.xlim([0,6.5])
plt.ylim([0,78])

plt.xlabel('Time, s', fontsize=20)
plt.ylabel('Taxels', fontsize=20)
plt.tight_layout(True)
#plt.show()
```

```python
# set writer
Writer = animation.writers['ffmpeg']
writer = Writer(fps=50, metadata=dict(artist='Me'), bitrate=50000)
```

```python
def animate(i):
    #plt.eventplot(spike_trains[i])
    plt.eventplot(spike_trains_pos[i], color='r')
    plt.eventplot(spike_trains_neg[i], color='b')
```

```python
ani = matplotlib.animation.FuncAnimation(fig, animate, frames=325, repeat=True)
ani.save(Path(args.target_dir) / f'fake_tactile_plot_{args.obj}_{args.weight_class}_ratio2.mp4', writer=writer)
```
