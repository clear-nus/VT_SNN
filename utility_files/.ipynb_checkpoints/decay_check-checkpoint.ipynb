```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
from os import listdir
from os.path import isfile, join
from pathlib import Path
from scipy.signal import savgol_filter
```

```python
data_dir = "/home/crslab/Documents/jun16/jun16_dummy_z_warmup/"
target_dir = "/home/crslab/some_python_examples/clear_lab_stable/neutouch_diag_tools/jun16/jun16_dummy_z_warmup/"
#data_dir = "/home/crslab/Documents/jun8/jun8_gtech_trace_warmup/"
```

```python
# list the folder by created time sorting
os.chdir(data_dir)
tactile_files = filter(os.path.isfile, os.listdir(data_dir))
tactile_files = [os.path.join(data_dir, f) for f in tactile_files] # add path to each file
tactile_files.sort(key=lambda x: os.path.getmtime(x))
```

```python
if not os.path.exists(target_dir):
    os.mkdir(target_dir)
```

```python
import pickle
[right_map, left_map] = pickle.load(open('/home/crslab/some_python_examples/clear_lab_stable/neutouch_diag_tools/mappings.pkl', 'rb'))
```

```python
# THIS READ TACTILE CODE for new sensors
def read_tactile_file(tactile_path):
    """Reads a tactile file from path. Returns a pandas dataframe."""
    obj_path = Path(tactile_path)
    df = pd.read_csv(
        obj_path,
        delimiter=" ",
        names=["polarity", "cell_index", "timestamp"],
        dtype={'polarity': int, 'cell_index': int, 'timestamp': float}
    )
    return df
```

```python
# read all tactile
tact_data_dfs = []
for fname in tactile_files:
    print(fname)
    tact_data_dfs.append(read_tactile_file(fname))
```

```python
print('We have', len(tact_data_dfs), 'tactile data samples.')
```

```python
a = tact_data_dfs[0]
```

```python
a.cell_index <= 40
```

```python
all_lengths_right = []
all_lengths_left  = []

for i in range(len(tact_data_dfs)-1):
    a = tact_data_dfs[i]
    l = a[a.cell_index <= 40]
    r = a[a.cell_index > 40]
    all_lengths_right.append(len(r))
    all_lengths_left.append(len(l))
```

```python
fig, ax = plt.subplots(2, figsize=(15,12))

ax[0].plot(all_lengths_left)
yhat = savgol_filter(np.array(all_lengths_left), 11, 3)
ax[0].plot(yhat)
ax[0].set_title('Left')

ax[1].plot(all_lengths_right)
yhat = savgol_filter(np.array(all_lengths_right), 11, 3)
ax[1].plot(yhat)
ax[1].set_title('Right')

plt.tight_layout(True)
plt.savefig(target_dir + 'spike_counts_warmup.png', dip=300)
plt.show()
```

### Checking for dead cells

```python
big_df = pd.concat(tact_data_dfs)
```

```python
big_df.shape
```

```python
existing_cells, appearences = np.unique( big_df.cell_index, return_counts=True )
```

```python
len(existing_cells)
```

```python
must_have_taxels = []
for i in range(1,81):
    if not (i==17 or i==49):
        must_have_taxels.append(i)
```

```python
existing_cells, appearences
```

```python
n_threshold = len(tact_data_dfs)
len(tact_data_dfs)
```

```python
dead_cells = []
for cell_id, n in zip(existing_cells, appearences):
    if n <= n_threshold:
        dead_cells.append(cell_id)
```

```python
for cell_number in must_have_taxels:
    if not cell_number in list(existing_cells):
        dead_cells.append(cell_number)
```

```python
dead_cells = pd.Series(dead_cells)
print('There are', len(dead_cells), ' dead cells')
dead_cells_left = dead_cells[dead_cells < 40]
dead_cells_right = dead_cells[dead_cells >= 40]
```

```python
left_finger_dead_cells = dead_cells_left.map(left_map).values
right_finger_dead_cells = dead_cells_right.map(right_map).values
```

```python
left_finger_dead_cells
```

```python
right_finger_dead_cells
```

```python
import scipy.io as sio
```

```python
taxel_locations = np.asarray(sio.loadmat('/home/crslab/some_python_examples/clear_lab_stable/neutouch_diag_tools/finger_lut.mat')['L2'])
taxel_locations.shape
```

```python
left_finger_img = np.zeros_like(taxel_locations)
right_finger_img = np.zeros_like(taxel_locations)
```

```python
for i in range(40):
    if i in left_finger_dead_cells.tolist():
        left_finger_img[taxel_locations == i] = 1
    if i in right_finger_dead_cells.tolist():
        right_finger_img[taxel_locations == i] = 1
```

```python
plt.imshow(left_finger_img)
plt.tight_layout(True)
plt.savefig(target_dir + 'left_muted.png', dip=300)
plt.show()
```

```python
plt.imshow(right_finger_img)
plt.tight_layout(True)
plt.savefig(target_dir + 'right_muted.png', dip=300)
plt.show()
```
